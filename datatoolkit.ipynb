1. What is NumPy, and why is it widely used in Python ??

* N-dimensional array object ndarray for storing large datasets efficiently.

* Mathematical functions to operate on arrays without writing loops manually.

* Tools for linear algebra, statistics, Fourier transforms, etc.

2. How does broadcasting work in NumPy ??

* NumPy is a feature that allows arithmetic operations between arrays of different shapes without explicitly copying data.

* It works by automatically expanding the smaller array along missing dimensions so the shapes match for element-wise operations.

*If the dimensions are equal → They are compatible.

* If one dimension is 1 → That array is stretched (broadcast) along that dimension.

* If the dimensions differ and neither is 1 → They are incompatible (error).

3. What is a Pandas DataFrame ??

* A Pandas DataFrame is a two-dimensional, tabular data structure in Python provided by the Pandas library.

* Rows → represent records or observations.

* Columns → represent features or variables.

4. Explain the use of the groupby() method in Pandas ??

* The groupby() method in Pandas is used to split data into groups based on one or more columns, then apply a function (like sum, mean, count) to each group.

* Split → Divide the data into groups based on column(s).

* Apply → Apply a function (sum, mean, max, count, custom function, etc.) to each group.

* Combine → Merge the results back into a DataFrame or Series.

5. Why is Seaborn preferred for statistical visualizations ??

* Seaborn is preferred for statistical visualizations because it’s designed specifically to make complex statistical plots simple, attractive, and informative all while working seamlessly with Pandas.

6. What are the differences between NumPy arrays and Python lists ??

* NumPy Array: Stores elements of the same type (homogeneous)
* Python List: Can store elements of different types (heterogeneous)

7. What is a heatmap, and when should it be used ??

* A heatmap is a data visualization technique that uses color intensity to represent values in a matrix or table.
* Visualize correlations between variables (e.g., in a correlation matrix).
* Show patterns in large datasets (e.g., website traffic over days and hours).
* Highlight intensity or magnitude of values (e.g., sales performance across regions and months).
* Compare values across two dimensions (e.g., exam scores of students across subjects).

8. What does the term “vectorized operation” mean in NumPy ??

* In NumPy, a vectorized operation means performing an operation on an entire array at once, without using explicit Python loops.

* Instead of processing each element one by one, NumPy applies the operation to all elements simultaneously using highly optimized C code under the hood.

9.  How does Matplotlib differ from Plotly ??

* Matplotlib: Primarily static plotting (with some interactive support via mpl_interactive)
* Plotly: Fully interactive plotting (hover tooltips, zoom, drag, etc.)

10. What is the significance of hierarchical indexing in Pandas ??

* Hierarchical indexing also called MultiIndex in Pandas lets you have multiple levels of index labels on rows or columns.

* It’s significant because it allows you to work with higher-dimensional data in a 2D DataFrame while keeping it organized and easy to query.

11. What is the role of Seaborn’s pairplot() function ??

* The pairplot() function in Seaborn is used to visualize relationships between multiple variables in a dataset at once.

12. What is the purpose of the describe() function in Pandas ??

* The describe() function in Pandas is used to quickly generate summary statistics of a DataFrame’s numerical (and optionally categorical) columns.

* It’s extremely helpful for exploratory data analysis (EDA) because it gives a quick overview of the dataset’s distribution and spread.

13. Why is handling missing data important in Pandas ??

* Handling missing data in Pandas is crucial because real-world datasets often have incomplete or null values, and ignoring them can lead to errors, inaccurate analysis, or misleading results.

14. What are the benefits of using Plotly for data visualization ??

*  Interactive Visualizations, Wide Range of Plot Types, Web Integration, High-Quality Visuals, Easy Data Handling.

15. How does NumPy handle multidimensional arrays ??

* NumPy handles multidimensional arrays efficiently by providing fast element-wise operations, flexible indexing, and functions along axes, making it ideal for linear algebra, image processing, and scientific computing.

16. What is the role of Bokeh in data visualization ??

* Bokeh is a Python library for interactive and web-ready data visualization. Its main role is to create high-quality plots that can be embedded in web applications, dashboards, or notebooks, while allowing real-time interaction with the data.

17. Explain the difference between apply() and map() in Pandas ??

* map():
*Used with: Series (single column).
* Purpose: Apply a function, dictionary, or mapping element-wise to each value in a Series.
* Returns: A new Series with transformed values.
* apply():
* Used with: Series or DataFrame.
* Purpose: Apply a function to each element, row, or column.
* Returns: Series (if applied to Series) or Series/DataFrame (if applied to DataFrame).

18. What are some advanced features of NumPy ??

* Broadcasting, Vectorized Operations,  Advanced Indexing, Linear Algebra Functions, Random Number Generation.

19. How does Pandas simplify time series analysis ??

* andas is extremely powerful for time series analysis because it provides specialized data structures and tools that make working with dates, times, and sequences much simpler.

20. What is the role of a pivot table in Pandas ??

* A pivot table in Pandas is used to summarize and reorganize data to make it easier to analyze. It’s similar to Excel’s pivot table feature.

* It allows you to aggregate data based on one or more keys (rows and columns) and compute summary statistics like sum, mean, count, etc.

21. Why is NumPy’s array slicing faster than Python’s list slicing ??

* NumPy’s array slicing is faster than Python’s list slicing because of how NumPy stores and accesses data internally.

```python
# A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
# Answer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math': [85, 78, 92, 65, 89],
    'Science': [88, 74, 95, 70, 85],
    'English': [75, 80, 85, 60, 78],
    'History': [70, 68, 82, 65, 80]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
# Generate a bar plot using Plotly
# Answer
import plotly.express as px
import pandas as pd

data = {
    'Fruit': ['Apples', 'Oranges', 'Bananas', 'Grapes', 'Mangoes'],
    'Sales': [120, 90, 150, 80, 200]
}

df = pd.DataFrame(data)

fig = px.bar(df,
             x='Fruit',
             y='Sales',
             title='Fruit Sales Data',
             labels={'Sales': 'Number of Sales', 'Fruit': 'Fruit Type'},
             color='Fruit')

fig.show()
```

```python
# Create a line plot with multiple lines using Matplotlib
# Answer
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 5, 6]

plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='s')
plt.plot(x, y3, label='Line 3', marker='^')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot Example')

plt.legend()

plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
# Answer
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 32, 28, 40]
}

df = pd.DataFrame(data)

filtered_df = df[df['Age'] > 30]

print("Original DataFrame:\n", df)
print("\nFiltered DataFrame (Age > 30):\n", filtered_df)
```

```python
# Perform matrix multiplication using NumPy
# Answer
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result1 = np.dot(A, B)

result2 = A @ B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("\nMatrix multiplication using np.dot():\n", result1)
print("\nMatrix multiplication using @ operator:\n", result2)
```

```python
# Create a 3D scatter plot using Plotly
# Answer
import plotly.express as px
import pandas as pd

data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 11, 12, 13, 14],
    'Z': [5, 6, 7, 8, 9],
    'Category': ['A', 'B', 'A', 'B', 'A']
}

df = pd.DataFrame(data)

fig = px.scatter_3d(df,
                    x='X',
                    y='Y',
                    z='Z',
                    color='Category',
                    size='Z',
                    title='3D Scatter Plot Example')

fig.show()
```
